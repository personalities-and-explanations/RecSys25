```python


stemmer = PorterStemmer()


def stem_phrase(phrase):
    return " ".join([stemmer.stem(w) for w in phrase.lower().split()])

trait_terms_stemmed = {
    trait: [stem_phrase(word) for word in words]
    for trait, words in trait_terms.items()
}


def process_explanation_words(df, translation_dict, trait_terms_stemmed):

    df["Word_EN"] = df["Word"].map(translation_dict)


    df["Word_EN_Stemmed"] = df["Word_EN"].apply(lambda w: stemmer.stem(w.lower().strip()) if pd.notna(w) else w)


    def match_trait_stemmed(word):
        if pd.isna(word): return None
        stemmed = stemmer.stem(word.lower().strip())
        for trait, phrases in trait_terms_stemmed.items():
            for phrase in phrases:
                if stemmed in phrase.split():
                    return trait
        return None

    df["Matched_Trait_Stemmed"] = df["Word_EN_Stemmed"].apply(match_trait_stemmed)


    total = len(df)
    matched_counts = df["Matched_Trait_Stemmed"].value_counts().to_dict()
    ground_truth_counts = {trait: len(words) for trait, words in trait_terms_stemmed.items()}

    results = []
    for trait in trait_terms_stemmed:
        matched = matched_counts.get(trait, 0)
        ground_truth = ground_truth_counts.get(trait, 1)
        precision = matched / total
        recall = matched / ground_truth
        f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
        results.append({
            "Trait": trait,
            "Precision": round(precision, 3),
            "Recall": round(recall, 3),
            "F1-Score": round(f1, 3)
        })

    metrics_df = pd.DataFrame(results)
    return df, metrics_df
```

```python

datasets = {
    "Agreeableness": (top_words_agreeableness_explanation, translation_dict_agreeableness),
    "Neuroticism": (top_words_neuroticism_explanation, translation_dict),
    "Openness": (top_words_openness_explanation, translation_dict_openness),
    "Conscientiousness": (top_words_conscientiousness_explanation, translation_dict_conscientiousness),
    "Extraversion": (top_words_extroversion_explanation, translation_dict_extroversion)
}


all_processed_dfs = {}
all_metrics_dfs = {}

for trait, (df, translation_dict) in datasets.items():
    processed_df, metrics_df = process_explanation_words(df, translation_dict, trait_terms_stemmed)
    all_processed_dfs[trait] = processed_df
    all_metrics_dfs[trait] = metrics_df
    print(f"\n--- {trait} ---")
    print(metrics_df)

summary_metrics = pd.concat(
    all_metrics_dfs.values(),
    keys=all_metrics_dfs.keys()
).reset_index(level=1, drop=True).rename_axis("Explanation_Trait").reset_index()

print("\nCombined Metrics:\n")
print(summary_metrics)
```
